# TOPIC - HISTOGRAM COMPUTATION

Histogram allows to visualize the distribution of pixel intensities in an image.


**FUCNTION USED ->** 
gray_histogram = cv2.calcHist(images, channels, mask, histSize, ranges)


**5 PARAMETERS ->** 

1. images: List of image arrays you want to analyze (e.g. [gray] for a grayscale image).
2. channels: Index of the channel to compute the histogram for (e.g. [0] for grayscale).
3. mask: Optional mask to specify which region of the image to include (use None for the full image).
4. histSize: Number of bins in the histogram (e.g. [256] for full 0–255 intensity resolution).
5. ranges: The intensity value range to be considered (e.g. [0, 256] for grayscale pixel values).


**CONCEPT OF BINS ->**
Bins are groups that count how many pixels fall within specific intensity ranges in an image.

if we write [256] as histSize then it means Dear OpenCV, please give me 256 bins — one for each possible intensity value ->
        So:
        Bin 0 → counts how many pixels have value 0
        Bin 1 → counts how many pixels have value 1
        ...
        Bin 255 → counts how many pixels are pure white



NOTE - can also make a mask over an image and calculate histogram for a specific region

```python
# 1. grayscale histogram

import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread(r"C:\Users\hp\Downloads\sample photos\tree.jpg")
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

gray_histogram = cv.calcHist([gray], [0], None, [256], [0,256])  #pass image as a list

plt.figure()
plt.title("Grayscale histogram")
plt.xlabel("bins")
plt.ylabel(" no. of pixels")
plt.plot(gray_histogram)
plt.xlim([0,256])
plt.show()
```

```python
# 2. coloured histogram

import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread(r"C:\Users\hp\Downloads\sample photos\tree.jpg")

plt.figure()
plt.title("Grayscale histogram")
plt.xlabel("bins")
plt.ylabel(" no. of pixels")

colors=('b', 'g', 'r')
for i,col in enumerate(colors):
    coloured_histogram = cv.calcHist([img], [i], None, [256], [0,256])  #pass image as a list
    plt.plot(coloured_histogram, color=col)
    plt.xlim([0,256])

plt.show()
```
